```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

### data(x) Acidity of milk , data(y) Density of milk

```python
def readData(filename):
    df=pd.read_csv(filename)
    return df.values
x=readData('./linearX.csv')
y=readData('./linearY.csv')
x=x.reshape((99,))
y=y.reshape((99,))
plt.scatter(x,y)
plt.show()
x=x-x.mean()/(x.std())
```

## Linear Regression algorithm

```python
def hypothesis(theta,x):
    return theta[1]*x + theta[0]

def error(X,Y,theta):
    total_error=0
    m=x.shape[0]
    for i in range(m):
        total_error+=(Y[i]-hypothesis(theta,X[i]))**2
    return 0.5*total_error

def gradient(X,Y,theta):
    grad=np.array([0.0,0.0])
    m=X.shape[0]
    for i in range(m):
        grad[0] += -1*(Y[i]-hypothesis(theta,X[i]))
        grad[1] += -1*(Y[i]-hypothesis(theta,X[i]))*X[i]
    return grad

def gradientDescent(X,Y,learning_rate,maxItr):
    grad=np.array([0.0,0.0])
    theta=np.array([0.0,0.0])
    for i in range(maxItr):
        grad=gradient(X,Y,theta)
        theta[0]=theta[0]-learning_rate*grad[0]
        theta[1]=theta[1]-learning_rate*grad[1]
    return theta

theta=gradientDescent(x,y,learning_rate=0.001,maxItr=500)
print(theta[0],theta[1])
plt.scatter(x,y)
plt.plot(x,hypothesis(theta,x),color='r')
plt.show()
```
